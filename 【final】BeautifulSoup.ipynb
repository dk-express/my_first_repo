```python
### 久保さんのコードをもとに拡張
import re

in_list = False
current_issue = ''
current_target = ''
issue_list =[]
example_list = []
for obj in target.find_all(): 
    if obj.name == 'h2' and obj.get('id') == '機械学習事例一覧':
        in_list = True
    if obj.name == 'h2' and obj.get('id') == 'まとめ':
        in_list = False

    if in_list == False:
        continue
        
    if obj.name == 'h3':
        current_issue = obj.text[1:] 
    if current_issue == '':
        continue
        
    if obj.name == 'h4' and obj.text != '': #目的、タイトル出力
        current_target = obj.text
        print(f"■{current_issue}")
        print(current_target)
        issue_list.append([current_issue, current_target])
        
    if obj.name == 'a': #事例のURL出力
        a1 = obj.string
        a2 = obj["href"]
        print(a1)
        print(a2, end = "\n\n")

    if obj.name == 'p' and obj.text.startswith('■ 活用技術：'):
        obj_ = obj
        exp = r'■ 活用技術：(.*)\n■ データ：(.*)\n■ 期待できる効果：(.*)'
        res = re.match(exp, obj.text)
        print(res.group(1))
        print(res.group(2))
        print(res.group(3))
        example_list.append([current_issue, current_target, a1, a2])    
    

    
```
